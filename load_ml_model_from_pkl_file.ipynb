```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import joblib

# Load the trained model
model = joblib.load('meachine_learning_models/sample_model_xgboost.pkl')

# New sample data (Assuming a single new claim record)
new_data = pd.DataFrame([{
    "AccountNumber": 123456,
    "VisitID": "V98765",
    "AdmitDate": "2025-03-10",
    "DischargeDate": "2025-03-15",
    "Name": "John Doe",
    "PatientID": "P1234",
    "LocationID": "L001",
    "AccountType": "Inpatient",  # Categorical feature
    "CurrentPrimaryInsurance": "Aetna",
    "CurrentFinancialClass": "Class A",
    "EntryBy": "Admin",
    "EntryDate": "2025-03-10",
    "DS_YEAR": 2025,
    "DS_MONTH": 3,
    "DS_DAY": 10,
    "DS_WD": 2,
    "Notes": "Patient admitted for surgery",
    "TransCount": 5,
    "TransAmount": 1500,
    "TransactionInsurance": "Yes",
    "PSTAmount": 200,
    "ADJAmount": 50,
    "PAYAmount": 1200,
    "REVAmount": 150,
    "BillNo": "B987654",
    "BillStatus": "Pending",  # Categorical feature
    "BillType": "Emergency",  # Categorical feature
    "BillFromDate": "2025-03-10",
    "BillThroughDate": "2025-03-15",
    "Reference1": "Ref1234",
    "Reference2": "Ref5678",
    "ClaimID": "C9876",
    "RMCCode": "RMC123",
    "ClaimGroupCode": "Group01",
    "ClaimDeptResponsible": "Surgery",
    "UpdateDateTime": "2025-03-10 12:30:00"
}])

# Label Encoding for categorical columns
label_encoder = LabelEncoder()
new_data['AccountType'] = label_encoder.fit_transform(new_data['AccountType'])
new_data['CurrentPrimaryInsurance'] = label_encoder.fit_transform(new_data['CurrentPrimaryInsurance'])
new_data['TransactionInsurance'] = label_encoder.fit_transform(new_data['TransactionInsurance'])
new_data['BillStatus'] = label_encoder.fit_transform(new_data['BillStatus'])
new_data['BillType'] = label_encoder.fit_transform(new_data['BillType'])

# Convert columns to numeric and handle NaN
numerical_columns = ['TransCount', 'TransAmount', 'PSTAmount', 'ADJAmount', 'PAYAmount', 'REVAmount']
new_data[numerical_columns] = new_data[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Handle missing values (if any)
new_data[numerical_columns].fillna(0, inplace=True)

# Scale the numeric columns (assuming StandardScaler was used during training)
scaler = StandardScaler()  # You can load your original scaler if you have it
new_data[numerical_columns] = scaler.fit_transform(new_data[numerical_columns])

# Make the prediction using the trained model
prediction = model.predict(new_data)

# Display the result
if prediction[0] == 0:
    print("The claim is NOT delayed.")
else:
    print("The claim is delayed.")
```
